### WHAT IS FIFA WORLD CUP?

The FIFA World Cup, often simply called the World Cup, is an international association football competition contested by the senior men's national teams of the members of the Fédération Internationale de Football Association (FIFA), the sport's global governing body. 

The championship has been awarded every four years since 1930, except in 1942 and 1946 when it was not held because of the Second World War.

```python
winner = winners["Winner"].value_counts()
runners_up = winners["Runners-Up"].value_counts()
third = winners["Third"].value_counts()
fourth = winners["Fourth"].value_counts()
winner
```

```python
index = set(winner.index.to_list() + runners_up.index.to_list() + third.index.to_list())
country = pd.DataFrame(index= index)
country = country.merge(winner, how='left',left_index=True, right_index=True)
country = country.merge(runners_up, how='left',left_index=True, right_index=True)
country = country.merge(third, how='left',left_index=True, right_index=True)
country = country.merge(fourth, how='left',left_index=True, right_index=True)
country.fillna(0,inplace = True)
country = country.sort_values(by=['Winner','Runners-Up','Third'],ascending =False)
country
```

```python
# DATA MANIPULATION 
# TOP teams score most goals 

home_goals = matches[['Home Team Name','Home Team Goals']]
away_goals = matches[['Away Team Name','Away Team Goals']]

home_goals = home_goals.rename(columns = {'Home Team Name':'Team', 'Home Team Goals':'Goals_home'})
away_goals = away_goals.rename(columns = {'Away Team Name':'Team', 'Away Team Goals':'Goals_away'})

home_goals = home_goals.groupby('Team', as_index= False).sum()
away_goals = away_goals.groupby('Team', as_index= False).sum()

team_goals = pd.merge(home_goals, away_goals, on = 'Team', how= 'right')

team_goals = team_goals.fillna(value=0)
team_goals['Goals_home'] = team_goals['Goals_home'].astype('int')

def teamgoal(datainput):
  x = datainput[0]
  y = datainput[1]
  return x + y

team_goals['Goals'] = team_goals[['Goals_home', 'Goals_away']].apply(teamgoal, axis=1)

team_goals_15 = team_goals.sort_values(by ='Goals', ascending = False).head(15)[['Team','Goals']]
team_goals_15
```

```python
#  Visualization - Goal _ team
# TOP teams score most goals in World Cup history

team_goals_15

fig, ax = plt.subplots(figsize=(12,7))
ax.bar(team_goals_15['Team'],team_goals_15['Goals'], width = 0.8)

ax.set(xlabel = "Team",
       ylabel = "Number of Goals ",
      title = "Top 15 teams scored most in World Cup history",
       xticks = team_goals_15['Team']
      )
ax.set_xticklabels(team_goals_15['Team'], rotation='39')

plt.show()
```

```python
# DATA MANIPULATION

# build function showing their name - team
def player_country(inputdata):
    x = inputdata[0]
    y = inputdata[1]
    return f'{x} - {y}'

# 15 Players participate in most game 
# Player plays the most in the starting of games 
players.replace({'Player Name':'M�LLER'}, {'Player Name':'Müller'},inplace = True)

# # Creating column Players in Years 
player_year_1 = players[['MatchID', 'Team Initials', 'Player Name']]
player_year_1['Player - Team'] = player_year_1[['Player Name', 'Team Initials']].apply(player_country, axis=1)
player_year_2 = player_year_1.rename(columns = {'MatchID':'Matches played'})
player_most = player_year_2[['Player - Team', 'Matches played']].groupby('Player - Team', as_index= False).count().sort_values(by='Matches played', ascending = False).head(15)

player_most
```

```python
# DATA MANIPULATION
# Player in year 
player_year_1
matches_year = matches[['Year', 'MatchID']]
number_player_year = pd.merge(player_year_1,matches_year, on= 'MatchID', how='right')

number_player_year = number_player_year[['Player - Team', 'Year']]
number_player_year = number_player_year.drop_duplicates(subset=['Player - Team', 'Year'], keep='first', inplace= False)
number_player_year = number_player_year.groupby('Year', as_index=False).count()
```

```python
# Visualization  - Player

# # Players participate in each tournament  
number_player_year

plt.figure(figsize=(15, 8)) # correct the size of the graph

plt.plot(number_player_year['Year'],number_player_year['Player - Team'], linestyle='solid')

plt.xlabel('Years')
plt.ylabel('Number of Players')

plt.title('Number of Player participating in World Cup Tournaments')

plt.show()
```

```python
# DATA MANIPULATION
# TOP 15 matches with highest attendence

def venue_game(inputdata):
    x = inputdata[0]
    y = inputdata[1]
    z = inputdata[2]
    return f'{z}: {x} vs {y}'

matches_fan = matches[['Year', 'Stadium', 'City', 'Attendance', 'Home Team Name', 'Away Team Name']]
matches_fan['Competition'] = matches_fan[['Home Team Name', 'Away Team Name', 'Year']].apply(venue_game, axis = 1)
matches_fan_15 = matches_fan.sort_values(by='Attendance', ascending= False)
matches_fan_15 = matches_fan_15[['Attendance', 'Competition']].head(15)
matches_fan_15
```

```python
# Visualization  - Matches: Attendance
# TOP 15 matches with highest attendance

matches_fan_15
fig, ax = plt.subplots(figsize=(15, 8))
ax.barh(matches_fan_15['Competition'], matches_fan_15['Attendance'])
plt.xlabel('Attendance')
plt.title('Top 15 matches with highest Attendance')
plt.show()
```
